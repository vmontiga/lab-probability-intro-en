## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

$$P(A) = \frac{\text{favorable outcomes}}{\text{total outcomes}}$$

Taking a white ball out.
$$ P(W) =  \frac{10}{30} = \frac{1}{3} $$

Taking a white ball (W) on first extraction and a black ball (B) out on second extraction without replacing the first one.

$$ P(W\cap B) =  P(W) \times P(B) = \frac{10}{30} \times \frac{8}{29} = \frac{8}{87} $$

Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.

$$ P(W\cap B) =  P(W) \times P(B) = \frac{10}{30} \times \frac{8}{30} = \frac{4}{45} $$

Taking 3 balls and them being all different colours.

Counting all possibilities:

$$ P(W\cap R \cap B) =  P(W\cap R \cap B) + P(W\cap B \cap R) + P(R\cap W \cap B) + P(R\cap B \cap W) + P(B\cap R \cap W) + P(B\cap W \cap R) $$

Since the probability of each draw of 3 balls is the same, we can simplify:

$$ P(W\cap R \cap B) =  6 \times P(W\cap R \cap B) $$

$$ P(W\cap R \cap B) =  6 \times (\frac{10}{30} \times \frac{12}{29} \times \frac{8}{28}) $$

$$ P(W\cap R \cap B) =  6 \times (\frac{8}{203}) $$

$$ P(W\cap R \cap B) =  \frac{48}{203} $$

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy. 
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

$ P(Cloudy) = 0.4 $
$ P(Rain|Cloudy) = 0.5 $
$ P(Rain|Non-Cloudy) = 0.2 $

$$ P(Rain) =  P(Rain|Cloudy)\times P(Cloudy) + P(Rain|Non-Cloudy) \times P(Non-Cloudy) $$

We can rewrite $P(Non-Cloudy) = 1 - P(Cloudy) = 0.6 $ since, by definition, all other days are not cloudy.

$$ P(Rain) =  0.5 \times 0.4 + 0.2 \times 0.6 $$

$$ P(Rain) =  0.32 $$

## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

* $P(\overline{A}) = 0.7$   -> **TRUE**

$$P(\overline{A}) = 1 - P(A) = 1 - 0.3 = 0.7$$

* $P(A \cup B) = 0.6$   -> **FALSE**

$$P(A \cup B) = P(A) + P(B) - P(A \cap B) = 0.3 + 0.6 - 0.1 = 0.8$$

* $P(A \cap \overline{B}) = 0.2$ -> **TRUE**

$$P(A \cap \overline{B}) = P(A) - P(A\cap B) = 0.3 - 0.1 = 0.2$$

* $P(A \cup \overline{A}) = 1$  -> **TRUE**

Because Either A or everything else than A happens, in other words it is the whole space therefore always 1.

* $P(A \cap (B \cup \overline{B})) = 0.4$ -> **TRUE**

We can simplify $ (B \cup \overline{B}) $ since it represents the whole space itself (previous point). Therefore it becomes just the probability of A:

$$P(A \cap (B \cup \overline{B})) = P(A) = 0.4$$

* $P(\overline{A} \cap \overline{B}) = 0.3$ -> **FALSE**

Since it represents neither A or B we can rephrase it as everything except either A, B or both ($P(A \cup B)$)
$$P(\overline{A} \cap \overline{B}) = 1 - P(A \cup B) = 1 - 0.8 = 0.2$$

* $P(\overline{A \cap B}) = 0.9$ -> **TRUE**

$$P(\overline{A \cap B}) = 1 - P(A\cap B) = 1 - 0.1 = 0.9$$

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

$P(Covid|China) = 0.01$

$P(Covid|Italy) = 0.2$

$P(China) = \frac{5}{6}$

$P(Italy) = \frac{1}{6}$



$$P(China|Covid) = \frac{P(Covid|China)\times P(China)}{P(Covid)}$$

From the script we can determine $P(Covid)$:

$P(Covid) = P(Covid|China) \times P(China) + P(Covid|Italy) \times P(Italy)$

$ P(Covid) = 0.01 \times \frac{5}{6} + 0.2 \times \frac{1}{6} = \frac{1}{24}$

$$P(China|Covid) = \frac{0.01 \times \frac{5}{6}}{\frac{1}{24}} = \frac{1}{5}$$

## BONUS

You roll 2 dice. What is the probability that the sum of the numbers is 7? And 8? And 9? Then simulate 10000 times and draw a countplot.

your explanation here

D1 = First Dice result

D2 = Second Dice result

Total outcomes = 6*6 = 36

S = D1 + D2 (Summ of the results)

P(S=7) combinations :  1+6 / 2+5 / 3+4 / 6+1 / 5+2 / 4+3

$P(S=7) = \frac{6}{36} = \frac{1}{6}$

P(S=8) combinations :  2+6 / 3+5 / 4+4 / 5+3 / 6+2 

$P(S=8) = \frac{5}{36}$

P(S=9) combinations :  3+6 / 4+5 / 5+4 / 6+3 

$P(S=9) = \frac{4}{36} = \frac{1}{9}$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

d1 = np.random.randint(1,7, 10000)
d2 = np.random.randint(1,7, 10000)

roll_summ = d1 + d2
# Set summ on x, the count goes on y
sns.countplot(x=roll_summ, palette = 'coolwarm')
plt.title("Distribution of the Rolls (10 000 random rolls)")
plt.ylabel('Count')
plt.xlabel('Sum')
plt.show()
```
